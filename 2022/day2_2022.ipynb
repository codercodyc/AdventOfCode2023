```python

scores = {'A': 1, 'B': 2, 'C': 3, 'loss': 0, 'tie': 3, 'win': 6}

outcomes = {
    'A': {'A': 'tie', 'B': 'loss', 'C': 'win'},
    'B': {'A': 'win', 'B': 'tie', 'C': 'loss'},
    'C': {'A': 'loss', 'B': 'win', 'C': 'tie'}}

def parse_data (data):

    # replace XYZ with ABC
    translation = str.maketrans('XYZ', 'ABC')
    data = data.translate(translation)

    data = data.split('\n')
    data = [tuple(x.split()) for x in data]
    return data

def round_score(round):
    pick_score = scores[round[1]]
    outcome = scores[outcomes[round[1]][round[0]]]
    return pick_score + outcome
    


scores = [round_score(round) for round in parse_data(data)]
print(scores)
sum(scores)
```
