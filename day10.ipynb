### Pipe Definitions
- `|` is a vertical pipe connecting north and south.
- `-` is a horizontal pipe connecting east and west.
- `L` is a 90-degree bend connecting north and east.
- `J` is a 90-degree bend connecting north and west.
- `7` is a 90-degree bend connecting south and west.
- `F` is a 90-degree bend connecting south and east.
- `.` is ground; there is no pipe in this tile.
- `S` is the starting position of the animal; there is a pipe on this tile, but your sketch doesn't show what shape the pipe has.

```python
# %%time

pipe_directions = {'|': 'NS', '-' : 'EW', 'L' : 'NE', 'J' : 'NW', '7' :'SW', 'F' : 'SE', 'S' : 'NSEW'}
direction_translations = {'N' : (-1, 0), 'S' : (1, 0), 'E' : (0, 1), 'W' : (0, -1)} # tuple of (row, col) change
connections = {'N' : 'S', 'S' : 'N', 'E' : 'W', 'W' : 'E'}

dir_translation_map = str.maketrans('NSEW', 'SNWE')

def parse_data(data):
    lines = data.splitlines()
    # add first and last . to each row
    lines = ['.' + line + '.' for line in lines]
    # add top and bottom row of .
    lines = ['.' * len(lines[0])] + lines + ['.' * len(lines[0])]
    return lines


def find_connecting_pipes(pipe):
    row, col = pipe[0]
    pipe_char = pipe[1]
    directions = pipe_directions.get(pipe_char)
    translations = [(direction_translations[direction], direction) for direction in directions]
    connections = [((row + translation[0],col + translation[1]), direction) for translation, direction in translations]
    return connections


def is_connection(pipe, pipe_positions):    # input data is sturctured (row, col), direction
    loc, src_char = pipe
    dest_directions = pipe_positions[loc]
    if dest_directions != '.':
        dest_chars = str.translate(pipe_directions[dest_directions], dir_translation_map)
        if src_char in dest_chars:
            # print('found connection', loc, src_char, dest_chars)
            return True
    else:
        return False
    
def solve_1(data):
    parsed_data = parse_data(data)

    pipe_positions = {(row,col) : pipe for row, line in enumerate(parsed_data) for col, pipe in enumerate(line)}
    start_position = [(pos, pipe) for pos, pipe in pipe_positions.items() if pipe == 'S'][0]

    # print('pipe positions', pipe_positions)

    length = 0
    pipe_distances = {}

    current_pipes = [start_position]
    next_pipes = []

    while current_pipes:
        current_pipe = current_pipes.pop()
        # print('current', current_pipe)
        pipe_distances.update({current_pipe[0] : length})
        connections = find_connecting_pipes(current_pipe)
        # print(connections)
        valid_connections = [connection[0] for connection in connections if is_connection(connection, pipe_positions)]
        # print('valid connections', valid_connections, 'length', length)
        valid_next_pipes = [(connection, pipe_positions.get(connection)) for connection in valid_connections if connection not in pipe_distances.keys()]
        # print('next',valid_next_pipes)
        next_pipes.extend(valid_next_pipes)
        if len(current_pipes) == 0:
            # print('next', next_pipes, length)
            current_pipes = next_pipes
            length += 1
            next_pipes = []
    # print('pipe distances', pipe_distances)

    max_distance = max(pipe_distances.values())
    return max_distance, pipe_distances


# print(solve_1(sample_data2))
max_distance, distances = solve_1(data)
print(max_distance)
```
