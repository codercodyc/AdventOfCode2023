```python
j_replacements = {HandType.FOUR_OF_A_KIND.value: HandType.FIVE_OF_A_KIND.value,
                    HandType.FULL_HOUSE.value: HandType.FIVE_OF_A_KIND.value,
                    HandType.THREE_OF_A_KIND.value: HandType.FOUR_OF_A_KIND.value,
                    HandType.TWO_PAIR.value: HandType.FULL_HOUSE.value,
                    HandType.ONE_PAIR.value: HandType.THREE_OF_A_KIND.value,
                    HandType.HIGH_CARD.value: HandType.ONE_PAIR.value
}

def rank_hand_2(hand):
    return tuple([card_ranks_2[char] for char in hand])

def classify_hands_2(hands):
    classified_hands = [[], [], [], [], [], [], []]# one list for each hand type

    for hand_data in hands:
        hand = hand_data[0]

        # character counts in hand
        char_counts = {char: hand.count(char) for char in hand}
        values = char_counts.values()

        # print(char_counts)
        rank = rank_hand_2(hand)
        result = (hand_data, rank)

        joker = 'J' in hand
        j_count = char_counts.get('J',0)

        # classify sets
        if (5 in values):
            # print(hand, 'Five of a Kind')
            classified_hands[HandType.FIVE_OF_A_KIND.value].append(result)
        elif (4 in values):
            # print(hand, 'Four of a Kind')
            if (joker):
                classified_hands[j_replacements[HandType.FOUR_OF_A_KIND.value]].append(result)
            else:
                classified_hands[HandType.FOUR_OF_A_KIND.value].append(result)
        elif (3 in values and 2 in values):
            # print(hand, 'Full House')
            if (joker):
                classified_hands[j_replacements[HandType.FULL_HOUSE.value]].append(result)
            else:
                classified_hands[HandType.FULL_HOUSE.value].append(result)
        elif (3 in values):
            # print(hand, 'Three of a Kind')
            if (joker):
                classified_hands[j_replacements[HandType.THREE_OF_A_KIND.value]].append(result)
            else:
                classified_hands[HandType.THREE_OF_A_KIND.value].append(result)
        elif (2 in values and len(values) == 3):
            # print(hand, 'Two Pair')
            if (joker):
                if (j_count == 2):
                    classified_hands[HandType.FOUR_OF_A_KIND.value].append(result)
                else:
                    classified_hands[j_replacements[HandType.TWO_PAIR.value]].append(result)
            else:
                classified_hands[HandType.TWO_PAIR.value].append(result)
        elif (2 in values):
            # print(hand, 'One Pair')
            if (joker):
                classified_hands[j_replacements[HandType.ONE_PAIR.value]].append(result)
            else:
                classified_hands[HandType.ONE_PAIR.value].append(result)
        else:
            # print(hand, 'High Card')
            if (joker):
                classified_hands[j_replacements[HandType.HIGH_CARD.value]].append(result)
            else:
                classified_hands[HandType.HIGH_CARD.value].append(result)

    return classified_hands


classified_sample = classify_hands_2(parse_data(sample_data))
print(solve_part_1(classified_sample))


classified_hands_2 = classify_hands_2(parse_data(data))
# print(classified_hands_2)
solve_part_1(classified_hands_2)
```
