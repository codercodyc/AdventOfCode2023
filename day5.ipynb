```python
# test_data = maps_array[0][0]
# test_data
# test_seed = 1514493331
# test_seed = 79

def check_single_map(map_data, seed):
    for line in map_data:
        line_range = range(line[1], line[1] + line[2])

        if seed in line_range:
            # print('Yes')
            destination = seed - line[1] + line[0]
            return destination
        
    return seed


def check_seed(maps, seed_list):
    current_seed = seed_list
    for map in maps:
        current_seed = check_single_map(map, current_seed)
        # print(current_seed)
    return current_seed

# print(check_single_map(sample_maps_array[0], 55))

sample_results = [check_seed(sample_maps_array, int(seed)) for seed in sample_seeds.split()]
print(sample_results,'\tmin:', min(sample_results))

results = [check_seed(maps_array, int(seed)) for seed in seeds.split()]
print(results,'\nmin:', min(results))

david_results = [check_seed(david_maps_array, int(seed)) for seed in david_seeds.split()]
print('David:', david_results,'\nmin:', min(david_results))

```

```python
%%time
# import time

# start_time = time.time()

maps_array = [create_map_array(map_data) for map_data in maps]


def get_input_and_dest_ranges(map_data):
    src_ranges = [[item[1], item[2]] for item in map_data] # list of tuples (start, length)
    dest_ranges = [[item[0], item[2]] for item in map_data] # list of tuples (dest, length)
    # print(input_ranges, dest_ranges)
    return src_ranges, dest_ranges


def map_intervals(map_data, seed_ranges):
    seed_intervals = create_part_2_seeds(seed_ranges)
    map_ranges = [get_input_and_dest_ranges(map) for map in map_data]
    # print(map_ranges)

    current_intervals = seed_intervals
    # print('current',current_intervals)

    # for seed_range in seed_ranges:
    # int_start = seed_range[0]
    # int_end = seed_range[0] + seed_range[1]
    final_intervals = []

    for range_set in map_ranges:
        # print(current_intervals)

        # print('\nrange_set', current_intervals)
        new_intervals = []
        src_ranges, dest_ranges = range_set
        # print('src_ranges', src_ranges, dest_ranges)
        for i, current_range in enumerate(src_ranges):

            temp_intervals = []

            # print('scr_start', current_range[0])
            src_start = current_range[0]
            src_end = current_range[0] + current_range[1]
            dest_start = dest_ranges[i][0]
            # dest_end = dest_ranges[i][0] + dest_ranges[i][1]
            # print('src', (src_start, src_end))

            while current_intervals:
                intervals = current_intervals.pop()
                # print('interval', [])
                int_start = intervals[0]
                int_end = intervals[1]
                # print('interval', (int_start, int_end))
                # int_start = interval[0]
                # int_end = interval[1]
                # print('int_start', int_start, int_end)


            # print(current_range, src_start, src_end, dest_start, dest_end)

                # get left, mid, right ranges
                left = (int_start, min(int_end, src_start))
                mid = (max(int_start, src_start), min(int_end, src_end))
                right = (max(int_start, src_end), int_end)

                # print(left, mid, right)

                if left[1] > left[0]:
                    # print('left\t', left)
                    temp_intervals.append(left)
                if mid[1] > mid[0]:
                    # print('mid\t', mid)
                    new_intervals.append((mid[0]-src_start + dest_start, mid[1]-src_start + dest_start))
                if right[1] > right[0]:
                    # print('right\t', right)
                    temp_intervals.append(right)

            current_intervals = temp_intervals
            # print('current_intervals', current_intervals)
            # print(temp_intervals)
        current_intervals = new_intervals + temp_intervals

    # print('solution:',current_intervals)

    first_indexes = [item[0] for item in current_intervals]
    # print('\nsolution:',first_indexes)
    return min(first_indexes)



print('\nsample_solution:',map_intervals(sample_maps_array, sample_seeds))
print('\nsolution:',map_intervals(maps_array, seeds))
print('\ndavid_solution:',map_intervals(david_maps_array, david_seeds))

# print("\n--- %s s ---" % ((time.time() - start_time)))
```
