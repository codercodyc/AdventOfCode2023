```python

import math as m

# expand rows and columns that contain no galaxies
def parse_data(data):
    return data.splitlines()

parsed_data = parse_data(data)

def expand_data(data_array):
    # get lists of empty rows and column indexes
    empty_rows = [row for row, line in enumerate(data_array) if set(line) == set('.')]
    empty_cols = [col for col in range(len(data_array[0])) if set([line[col] for line in data_array]) == set('.')]
    # print('Empty:', empty_rows, empty_cols)

    # create sets of row and column indexes to add
    row_indexes = [row + index for index, row in enumerate(empty_rows)]
    col_indexes = [col + index for index, col in enumerate(empty_cols)]

    # print(len(data_array), len(data_array[0]))

    # add empty rows and columns
    for row in row_indexes:
        data_array.insert(row, '.' * len(data_array[0]))
    for col in col_indexes:
        for row, line in enumerate(data_array):

            data_array[row] = line[:col] + '.' + line[col:]

    return data_array


expanded_data = expand_data(parsed_data)


def get_galaxy_sets(data_array):
    positions = [(row,col) for row, line in enumerate(data_array) for col, char in enumerate(line) if char == '#']
    # print(positions, m.comb(len(positions), 2))
    galaxy_sets = []
    while positions:
        current_position = positions.pop(0)
        [galaxy_sets.append((current_position, position)) for position in positions]
    # print(len(galaxy_sets))
    return galaxy_sets

galaxy_sets = get_galaxy_sets(expanded_data)

def steps_between_galaxies(galaxy_set):
    pos1, pos2 = galaxy_set
    x1, y1 = pos1
    x2, y2 = pos2

    return abs(x1-x2) + abs(y1-y2)

def solve_1(galaxy_sets):
    steps = [steps_between_galaxies(set) for set in galaxy_sets]
    return sum(steps)

print(solve_1(galaxy_sets))
```
