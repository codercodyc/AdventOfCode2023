```python
def parse_data(data):
    data = data.splitlines()
    lr_table = str.maketrans('LR', '01')
    instructions = tuple(int(char.translate(lr_table)) for char in data[0])  # parse RL instructions
    nodes = [node.translate({ord(i): None for i in '=(),'}).split()for node in data[2:]]
    nodes = {node[0]: (node[1], node[2]) for node in nodes}
    return instructions, nodes

```

```python
start_node = 'AAA'
end_node = 'ZZZ'

def find_path(data):

    parsed_data = parse_data(data)
    # print(parsed_data)
    current_node = start_node
    count = 0

    while current_node != end_node:
            for i, instruction in enumerate(parsed_data[0]):
                current_node = parsed_data[1][current_node][instruction]
                count += 1
                # print(current_node)
    return count

find_path(data)
```
